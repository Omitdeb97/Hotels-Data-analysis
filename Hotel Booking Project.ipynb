# Data Analysis and Visualizations

```python
# Finding the reservation accertance and cancelation percentage rate (0: accept and 1 cancel)
canceled_percentage = df['is_canceled'].value_counts(normalize = True)
canceled_percentage
```

```python
# Finding the reservation accertance and cancelation percentage rate (0: accept and 1 cancel). Here is visualization
canceled_percentage = df['is_canceled'].value_counts(normalize = True)
print(canceled_percentage)

plt.figure(figsize = (5, 4))
plt.title('Reservation status count')
plt.bar(['Not cenceled', 'Cenceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.9)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor(1,1))
plt.title('Different hotels reservation status', size = 25)
plt.xlable('hotel')
plt.ylable("numver of reservations")
plt.show()
```

```python
resort_hotel = df[df['hotel'] =='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```
